```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sn
```

```python
df = pd.read_csv("covid-19.csv")
```

```python
df
```

```python
df=df.drop('Unnamed: 0',axis='columns')
```

```python
sn.heatmap(df.isnull(),cmap='viridis',yticklabels=False,cbar=False)
```

```python
df.shape
```

```python
df.describe()
```

```python
df=df.rename(columns={ "Confirmed":"confirm","Death": "death","newConfirmed":"newconfirm",'newDeath':'newdeath'})
df.head()
```

```python
country = df.groupby("Country")[['confirm' ,'death','newconfirm','newdeath']].sum().reset_index()
```

```python
country.head()
```

```python
import folium
worldmap = folium.Map(location=[32.4279,53.6880 ], zoom_start=4,tiles='Shavej Shaikh')

for Lat, Long, state in zip(33.93911,,country['Country']):
    folium.CircleMarker([Lat, Long],
                        radius=5,
                        color='red',
                      popup =('State: ' + str(state) + '<br>'),
                        fill_color='red',
                        fill_opacity=0.7 ).add_to(worldmap)
worldmap
```
