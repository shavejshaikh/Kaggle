```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("Prostate_Cancer.csv")
df.head()
```

```python
df=df.drop('id',axis=1)
```

```python
df.diagnosis_result =[1 if each == 'M' else 0 for each in df.diagnosis_result]
```

```python
df
```

## Find Null Values 

```python
sn.heatmap(df.isnull(),cmap='viridis',yticklabels=False,cbar=False)
```

## Exploratory Data Analysis 

```python
plt.plot(df['radius'],np.zeros_like(df['radius']),color='red')
plt.plot(df['texture'],np.zeros_like(df['texture']),color='green')
```

```python
sn.FacetGrid(df,hue='diagnosis_result',height=7).map(plt.scatter,'radius','texture').add_legend()
plt.show()
```

```python
sn.pairplot(df,hue='diagnosis_result')
```
