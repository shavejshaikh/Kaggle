```python
normalday=df[df['Final de Semana']==0.0]
weekend=df[df['Final de Semana']==1.0]
```

```python
weekend.head()
```

```python
plt.plot(normalday['Temperatura Media (C)'],np.zeros_like(normalday['Temperatura Media (C)']),'o')
plt.plot(weekend['Temperatura Media (C)'],np.zeros_like(weekend['Temperatura Media (C)']),'*',color='red')
```

```python
plt.plot(normalday['Consumo de cerveja (litros)'],np.zeros_like(normalday['Consumo de cerveja (litros)']),'o')
plt.plot(weekend['Consumo de cerveja (litros)'],np.zeros_like(weekend['Consumo de cerveja (litros)']),'*',color='red')
```

```python
plt.scatter(normalday['Consumo de cerveja (litros)'],normalday['Temperatura Minima (C)'],color='red')
plt.scatter(normalday['Consumo de cerveja (litros)'],normalday['Temperatura Media (C)'],color='black')
plt.scatter(normalday['Consumo de cerveja (litros)'],normalday['Temperatura Maxima (C)'])
```

### Bivariate Analyses

```python
plt.scatter(weekend['Temperatura Minima (C)'],weekend['Consumo de cerveja (litros)'],color='red')
plt.scatter(weekend['Temperatura Media (C)'],weekend['Consumo de cerveja (litros)'],color='black')
plt.scatter(weekend['Temperatura Maxima (C)'],weekend['Consumo de cerveja (litros)'])
```

```python
val=[]
n1=sum(normalday['Consumo de cerveja (litros)'])/5
val.append(n1)
n2=sum(weekend['Consumo de cerveja (litros)'])/2
val.append(n2)

my_labels ='WeekDays','Weekend'
plt.pie(val,labels=my_labels,autopct='%1.1f%%')
plt.title('Weekend Vs Weekdays')
plt.axis('equal')
plt.show()
```
