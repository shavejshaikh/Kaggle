```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('AirPassengers.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.plot(df.Month,df['#Passengers'],color='green')
plt.xlabel("Month")
plt.ylabel('Passengers')
plt.title("Month VS Passengers")
```

```python
plt.scatter(df.Month,df['#Passengers'],color='green')
plt.xlabel("Month")
plt.ylabel('Passengers')
plt.title("Month VS Passengers")
```

```python
df['Month']=pd.to_datetime(df['Month'])
```

```python
df.head()
```

```python
df.dtypes
```

```python
index_data=df.set_index('Month')
```

```python
index_data.head()
```

```python
plt.plot(index_data,color='blue')
plt.xlabel('Date')
plt.ylabel("Passengers")
```

```python
four_months_moving_average = index_data.rolling(window=4).mean()

plt.plot(index_data, color='blue', label='Original')

plt.plot(four_months_moving_average, color='red', label='Rolling Mean')

plt.legend(loc='best')
plt.title('4 Months Moving Average')
plt.show()
```

```python
six_months_moving_average = index_data.rolling(window=6).mean()

plt.plot(index_data, color='blue', label='Original')

plt.plot(six_months_moving_average, color='red', label='Rolling Mean')

plt.legend(loc='best')
plt.title('6 Months Moving Average')
plt.show()
```
